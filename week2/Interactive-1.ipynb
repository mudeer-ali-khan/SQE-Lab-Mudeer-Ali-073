```python
import numpy as np
import matplotlib.pyplot as plt

# -----------------------------
# DATA (YOUR EXCEL VALUES)
# -----------------------------
categories = [
    "Functional Suitability",
    "Performance Efficiency",
    "Compatibility",
    "Usability",
    "Reliability",
    "Security",
    "Maintainability",
    "Portability"
]

# App 1 (LinkedIn) — keep same as your report
APP1_RATINGS = [5, 3, 4, 4, 4, 4, 3, 5]

# App 2 (Indeed) — YOUR UPDATED EXCEL VALUES
APP2_RATINGS = [5, 4, 4, 4, 4, 4, 3, 5]

# Calculate average scores
APP1_SCORE = round(sum(APP1_RATINGS) / len(APP1_RATINGS), 2)
APP2_SCORE = round(sum(APP2_RATINGS) / len(APP2_RATINGS), 2)

# -----------------------------
# RADAR CHART SETUP
# -----------------------------
N = len(categories)
angles = np.linspace(0, 2 * np.pi, N, endpoint=False).tolist()

# Close the loop
APP1 = APP1_RATINGS + APP1_RATINGS[:1]
APP2 = APP2_RATINGS + APP2_RATINGS[:1]
angles += angles[:1]

# -----------------------------
# PLOT
# -----------------------------
plt.figure(figsize=(8, 8))
ax = plt.subplot(111, polar=True)

# Plot LinkedIn
ax.plot(angles, APP1, linewidth=2,
        label=f"LinkedIn (Score: {APP1_SCORE})")
ax.fill(angles, APP1, alpha=0.1)

# Plot Indeed
ax.plot(angles, APP2, linewidth=2, linestyle='dashed',
        label=f"Indeed (Score: {APP2_SCORE})")
ax.fill(angles, APP2, alpha=0.1)

# Labels
ax.set_xticks(angles[:-1])
ax.set_xticklabels(categories)

# Scale MUST be 1–5
ax.set_yticks([1, 2, 3, 4, 5])
ax.set_ylim(0, 5)

# Title
plt.title("Quality Comparison: LinkedIn vs Indeed", size=14)

# Legend
plt.legend(loc='upper right', bbox_to_anchor=(1.3, 1.1))

# Save chart
plt.savefig("radar_chart.png", dpi=300, bbox_inches='tight')

# Show chart
plt.show()
```
